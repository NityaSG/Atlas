```python
import pandas as pd
from sklearn.utils import shuffle

# Load the data from the CSV files
df1 = pd.read_csv('valid_data.csv')
df2 = pd.read_csv('invalid_data.csv')
df3 = pd.read_csv('rejected_df.csv')

# Concatenate the dataframes
df = pd.concat([df1, df2, df3])

# Shuffle the data
df = shuffle(df)

# Save the shuffled data to a new CSV file
df.to_csv('combined_shuffled.csv', index=False)
```

```python
import pandas as pd
from Bio import Entrez

# Read the CSV file
df = pd.read_csv('Integerised.csv')

# Function to fetch abstract
def fetch_abstract(pmid):
    Entrez.email = 'nityasingh030301@gmail.com'  # Always tell NCBI who you are
    handle = Entrez.efetch(db='pubmed', id=pmid, retmode='xml')
    article = Entrez.read(handle)['PubmedArticle'][0]
    abstract = article['MedlineCitation']['Article']['Abstract']['AbstractText'][0]
    return abstract

# Create a new DataFrame
new_df = pd.DataFrame(columns=['Reference ID', 'Article Classification', 'Abstract'])

for index, row in df.iterrows():
    try:
        abstract = fetch_abstract(row['Reference ID'])
        new_df = new_df.append({'Reference ID': row['Reference ID'], 'Article Classification': row['Article Classification'], 'Abstract': abstract}, ignore_index=True)
    except Exception as e:
        print(f"Error occurred while fetching abstract for Reference {row['Reference ID']}: {str(e)}")

# Save the new DataFrame to a CSV file
new_df.to_csv('abstracts.csv', index=False)
```
